Here’s our dataset description neatly separated by headings for clarity:

## Customer Information
- **CustomerID**: Unique customer ID  
- **Gender**: Gender of customer  
- **MaritalStatus**: Marital status of customer  
- **NumberOfAddress**: Total number of addresses added by the customer  

## Churn & Tenure
- **Churn**: Churn flag (whether customer has churned)  
- **Tenure**: Tenure of customer in the organization  

## Login & Device
- **PreferredLoginDevice**: Preferred login device of customer  
- **NumberOfDeviceRegistered**: Total number of devices registered by the customer  

## Location & Distance
- **CityTier**: City tier classification  
- **WarehouseToHome**: Distance between warehouse and customer’s home  

## Payment & Orders
- **PreferredPaymentMode**: Preferred payment method of customer  
- **PreferedOrderCat**: Preferred order category in the last month  
- **OrderAmountHikeFromlastYear**: Percentage increase in order amount compared to last year  
- **CouponUsed**: Total number of coupons used in the last month  
- **OrderCount**: Total number of orders placed in the last month  
- **DaySinceLastOrder**: Days since last order by customer  

## Engagement & Satisfaction
- **HourSpendOnApp**: Number of hours spent on mobile app or website  
- **SatisfactionScore**: Satisfaction score of customer on service  
- **Complain**: Whether any complaint was raised in the last month  
- **CashbackAmount**: Average cashback received in the last month
